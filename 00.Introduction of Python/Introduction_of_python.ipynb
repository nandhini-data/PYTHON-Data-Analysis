# <center>Python</center>

Python is a free open source programming language that is both powerful and easy to learn
It is widely used in many application
* Data Analytics
* Machine Learning
* Web Scraping

## Why python for analytics?

1. Scalability - Free to use,open source.
2. Versatility - With powerful libraries and framework used for data analysis and machine learning.
3. community - Large and active python user community.
4. Automation - Automate the complex task out of box.
5. Demand - Python skills are valuable and are becoming increasingly popular among analytics and BI professionals.

## Python analytics ecosystem

1. <span style='color:purple'>**General purpose programming language**</span><br>
    Mastering base **Python** will give foundation understanding of language which is essential for using packages and libraries.
2. <span style='color:purple'>**Data Manipulation and Analysis**</span><br>
   **Pandas** helps to structure the data into dataset format similar to Excel and sql.
3. <span style='color:purple'>**Data Visualization**</span><br>
    **Matplotlib** and **Seaborn** can create a wide array of visually appealing,static visualization.<br>
    **Plotly** can be used to create an interactive visualization and dynamic dashboard.
2. <span style='color:purple'>**Machine Learning**</span><br>
   **Scikit Learn** is used for building and testing machine learning models.<br>
   **Statsmodel** is used for model building and statistical analysis.<br>
   **Tensorflow** is the industry standard for developing deep learning models
